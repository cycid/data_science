```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_predict
```

```python
df=pd.read_csv('tit_train.xls', index_col='PassengerId')
test=pd.read_csv('tit_test.xls', index_col='PassengerId')
test.head(20)
```

```python
#replace nan in Age column
res=int(df['Age'].mean())
df['Age'].replace(np.nan, res, inplace=True)
df.head(20)

#for test dataset
test_av=int(test['Age'].mean())
test['Age'].replace(np.nan, res, inplace=True)
fare_mean=int(test['Fare'].mean())
test['Fare'].replace(np.nan, fare_mean, inplace=True)

test.head(20)
```

```python
clas=df.where(df['Pclass']==3)
clas.dropna(subset='Pclass', inplace=True)
clas.head(50)


#for test
```

```python
#   (<1)
df['Age']=df['Age'].apply(lambda x: x if x>=1 else x*100)
df.head(50)


#for test
test['Age']=test['Age'].apply(lambda x: x if x>=1 else x*100)
test.head(50)
```

```python
#     (male-1, female-0)
df['Sex']=df['Sex'].apply(lambda x: 0 if x=='female' else 1)
df.head(60)

#for test
test['Sex']=test['Sex'].apply(lambda x: 0 if x=='female' else 1)
test.head(60)
```

```python
col='Embarked'
le = LabelEncoder().fit(df[col].astype(str)) # convert to str first since le may fail due to difference type of data
df[col] = le.transform(df[col].astype(str))
df.head(20)


#for test

e = LabelEncoder().fit(test[col].astype(str)) # convert to str first since le may fail due to difference type of data
test[col] = e.transform(test[col].astype(str))
test.head(20)
```

```python
X=df.loc[:,['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
y=df.loc[:,'Survived']
X_train, X_test,y_train,y_test=train_test_split(X,y, test_size=0.25)


X_kaggle=test.loc[:,['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
```

```python
clf= MLPClassifier(
    activation='logistic',
    hidden_layer_sizes=(10,10,10),
    solver= 'lbfgs',
    verbose= True,
    random_state=0,
    alpha = 0.000001,
    max_iter=100000
)
clf.fit(X_train,y_train)


#cv_results=cross_validate(clf, X, y, return_estimator=True)
#cv_results
```

```python
print("train accuracy= {:.3%}".format(clf.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score (X_test, y_test)))
```
