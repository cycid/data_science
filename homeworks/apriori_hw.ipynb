```python
import pandas as pd
from apyori import apriori
from mlxtend.frequent_patterns import apriori, association_rules
import numpy as np
```

```python
df=pd.read_csv('Groceries_dataset.csv')
```

```python
df.head(10)
```

```python
#create true/false for every product in shop
after_dummies=df.itemDescription.str.get_dummies()
after_dummies[0:5]
```

```python
#concat two dataframes
df_v = pd.concat([df, after_dummies], axis=1)
df_v.shape
```

```python
list_of_columns=df_v.columns[3:170]
res=df_v.groupby(['Member_number','Date'])[list_of_columns].agg('sum')
res.head(10)

#b=res.ge(2).sum()
```
