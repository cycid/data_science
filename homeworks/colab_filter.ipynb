```python
from surprise import SVD
from surprise import Dataset, Reader
from surprise.model_selection import cross_validate
import pandas as pd
import surprise
```

```python
df=pd.read_excel('ratings_movies.xlsx', header=None)
df.head(20)
```

```python
min_val=df.min()
min_val=min_val.min()
max_val=df.max()
max_val=max_val.max()
```

```python
iid_list=[]
list_of_index=[]
list_of_data=[]
for index, row in df.iterrows():
    for value, column_name in zip(row.tolist(),df.columns):
        list_of_index.append(index)
        list_of_data.append(value)
        iid_list.append(column_name)
zipped=list(zip(list_of_index, iid_list, list_of_data))
data=pd.DataFrame(zipped, columns=['user_id', 'iid', 'score'])
data.head(20)
```

```python
#add user
user=pd.DataFrame({'user_id':[16,16,16], 'iid':[2, 5, 7], 'score':[1, 5, 5]})

data=pd.concat([data, user], axis=0, ignore_index=True)
```

```python
reader=Reader(rating_scale=(min_val,max_val))
data1=Dataset.load_from_df(data,reader)
data1.build_full_trainset()
```

```python
alg=surprise.SVDpp()
output = alg.fit(data1.build_full_trainset())
```

```python
pred= alg.predict(uid= '16', iid= '10')
pred, pred.est
```
